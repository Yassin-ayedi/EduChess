```python
from board_detction import get_squares
from get_position import get_fen_from_board
import cv2
import chess
import torch
from PIL import Image 

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")



```

```python
model_for_piece = torch.load(r'models/chess_piece_classifier.pth')  # Load the piece classification model
model_for_piece = model_for_piece.to(DEVICE)  

# Load the color model
model_for_color = torch.load(r'models/chess_color_classifier.pth')  # Load the color classification model
model_for_color = model_for_color.to(DEVICE) 
```

```python
image = cv2.imread(r"images/1-e4.JPG")
info=get_squares(image,True)


fen=get_fen_from_board(info,model_for_piece,model_for_color)
```
